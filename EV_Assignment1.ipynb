```python
import matplotlib.pyplot as plt
import seaborn as sns
#Distribution of EV Range
plt.figure(figsize=(8,5))
sns.histplot(df['range_km'], kde=True)
plt.title('Distribution of EV Range')
plt.xlabel('Range (km)')
plt.ylabel('Count')
plt.show()
```

```python
#Top 10 EV brands by count
plt.figure(figsize=(10,5))
df['brand'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 EV Brands')
plt.xlabel('Brand')
plt.ylabel('Number of Models')
plt.show()
```

```python
# Range vs Battery Capacity
plt.figure(figsize=(8,5))
sns.scatterplot(x=df['battery_capacity_kWh'], y=df['range_km'])
plt.title('Battery Capacity vs Range')
plt.xlabel('Battery Capacity (kWh)')
plt.ylabel('Range (km)')
plt.show()
```

```python
#Top Speed Distribution
plt.figure(figsize=(8,5))
sns.histplot(df['top_speed_kmh'], kde=True)
plt.title('Top Speed Distribution')
plt.xlabel('Top Speed (km/h)')
plt.ylabel('Count')
plt.show()
```

```python
#Correlation Heatmap
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
